# Import / Input

The following code will import the [pandas](https://pandas.pydata.org/) code library, import player statistics for the Toronto Raptors 2023 season, then select just the 'Player', 'FG%' (Field Goal Percentage), and 'FT%' (Free Throw Percentage) columns.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/pbeens/Data-Dunkers/main/Data/raptors-2023.csv'
raptors_df = pd.read_csv(url)

raptors_df = raptors_df[['Player', 'FG%', 'FT%']]

display(raptors_df)
```

# Process

## Add a Column

Calculate the shot average for each player and add it as a new column to the DataFrame. The 'Shot Average (%)' is calculated as the average of 'FG%' and 'FT%', multiplied by 100.

```python
raptors_df['Shot Average (%)'] = (raptors_df['FG%'] + raptors_df['FT%']) / 2 * 100

display(raptors_df)
```

## Rounding

Round the 'Shot Average (%)' column to one decimal place for clarity.

```python
raptors_df['Shot Average (%)'] = raptors_df['Shot Average (%)'].round(1)

display(raptors_df)
```
