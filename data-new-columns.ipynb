# Import / Input

```python
# Import the pandas library
import pandas as pd

# Set the URL of the data file
# The URL points to the CSV file containing player statistics for the Toronto Raptors 2023 season.
url = 'https://raw.githubusercontent.com/pbeens/Data-Dunkers/main/Data/raptors-2023.csv'

# Read the data file into a pandas DataFrame named 'raptors_df'
# Load the CSV data from the URL into a DataFrame for processing.
raptors_df = pd.read_csv(url)

# Select specific columns from the DataFrame
# We are interested in the 'Player', 'FG%' (Field Goal Percentage), and 'FT%' (Free Throw Percentage) columns.
raptors_df = raptors_df[['Player', 'FG%', 'FT%']]

display(raptors_df)
```

# Process

## Add a Column

```python
# Calculate the shot average for each player and add it as a new column to the DataFrame
# The 'Shot Average (%)' is calculated as the average of 'FG%' and 'FT%', then multiplied by 100.
raptors_df['Shot Average (%)'] = (raptors_df['FG%'] + raptors_df['FT%']) / 2 * 100

# Display the DataFrame
display(raptors_df)
```

## Rounding

```python
# Round the 'Shot Average (%)' column to one decimal place for clarity.
raptors_df['Shot Average (%)'] = raptors_df['Shot Average (%)'].round(1)

# Display the DataFrame
display(raptors_df)
```
