# Import / Input

```python
# import plotly.express and pandas
import plotly.express as px
import pandas as pd

# Read the CSV file into a DataFrame named df
url = 'https://raw.githubusercontent.com/pbeens/Data-Dunkers/main/Data/x-y-data.csv'
df = pd.read_csv(url)
```

# Process (Head/Tail/Columns)

```python
# Head
# Prints 5 rows if not specified

print(df.head())
```

```python
# Display the last 2 rows of the data

print(df.tail(2))
```

```python
# Display the column names

print(df.columns)
```

```python
# Changing column names

df.columns = ['X Value', 'X^2']
print(df.columns)
```

```python
# Preferred renaming method with lots of columns
# (This changes the column names back to the original)

df = df.rename(columns={'X Value': 'X',
                        'X^2': 'Y'})
print(df.columns)
```

# Output

```python
# Create the plot
fig = px.line(data_frame=df,
              x='X',
              y='Y',
              title='Data from a CSV file')

# Show the plot
fig.show()
```
