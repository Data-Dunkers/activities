# Import / Input

```python
import plotly.express as px
import pandas as pd

# Read the CSV file into a DataFrame named df
url = 'https://raw.githubusercontent.com/pbeens/Data-Dunkers/main/Data/x-y-data.csv'
df = pd.read_csv(url)

df
```

# Process (Head/Tail/Columns)

Display the first 5 rows.

```python
df.head()
```

Display the last 2 rows.

```python
df.tail(2)
```

Display the column names.

```python
df.columns
```

Change the column names.

```python
df.columns = ['X Value', 'X^2']

df.columns
```

For renaming when we have lots of columns we can use `.rename(columns={'column name':'new column name'})`

In this example we are changing them back to their original values.

```python
df = df.rename(columns={'X Value':'X', 'X^2':'Y'})

df.columns
```

# Output

Create a line graph from the dataframe.

```python
px.line(df, x='X', y='Y', title='Data from a CSV file')
```
