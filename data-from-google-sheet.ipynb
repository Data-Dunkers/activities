# Import / Input

When importing from a Google Sheet, we need to change the end of the url to `/export?format=csv` and then read it like a CSV file.

```python
import pandas as pd
import plotly.express as px

url = 'https://docs.google.com/spreadsheets/d/1ZULKhYzsMd4eYwiprsyGgE9Df3gaVtO8WRalUQDn-xE/export?format=csv'

df = pd.read_csv(url)
```

# Output

Create the plot. This code is the same as the CSV example.

```python
px.line(df, x='X', y='Y', title='Data from a CSV file')
```
