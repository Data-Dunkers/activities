# Import / Input

Read the Excel file into a DataFrame named df.

```python
import plotly.express as px
import pandas as pd

url = 'https://raw.githubusercontent.com/pbeens/Data-Dunkers/main/Data/x-y-data.xlsx'
df = pd.read_excel(url)
```

# Output

Create the plot. This code is the same as the CSV example.

```python
px.line(df, x='X', y='Y', title='Data from a CSV file')
```
