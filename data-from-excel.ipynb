# Import / Input

```python
# import plotly.express and pandas
import plotly.express as px
import pandas as pd

# Read the Excel file into a DataFrame named df
url = 'https://raw.githubusercontent.com/pbeens/Data-Dunkers/main/Data/x-y-data.xlsx'
df = pd.read_excel(url)
```

# Output

```python
# Create the plot
# This code is the same as the CSV example!

fig = px.line(data_frame=df,
              x='X',
              y='Y',
              title='Data from a CSV file')

# Show the plot
fig.show()
```
