# Import / Input

```python
import pandas as pd

# URL of the CSV file containing data for Pascal Siakam
url = 'https://raw.githubusercontent.com/pbeens/Data-Dunkers/main/Data/Pascal_Siakam.csv'

# Read the CSV file into a pandas DataFrame named df
df = pd.read_csv(url)

# Display the DataFrame
display(df)
```

# Process

```python
# Drop the index 9 row, then display the updated dataframe

df = df.drop(9)
display(df)
```

![](https://raw.githubusercontent.com/pbeens/Images/main/other/dd-presentation/filter_by_season.png)

```python
# Filter the DataFrame to exclude seasons after 2022-23
filter = df['SEASON_ID'] <= '2022-23'
df = df[filter]

# Display the updated dataframe
display(df)
```

```python
# Apply a compound filter
# Q. Does this filter update the dataframe?

filter_1 = df['FT_PCT'] > 0.75  # free throw % above 75%
filter_2 = df['GS'] == df['GP']  # games started equals games played

# Display the filtered content
display(df[filter_1 & filter_2])
```
